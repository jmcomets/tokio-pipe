```python
import csv
import matplotlib.pyplot as plt

def plot_csv(filename, delimiter=';', plt=plt,
             title=None, read_labels_from_first_line=True,
             xtype=int, xlabel=None, xindex=0,
             ytype=float, ylabel=None, yindex=1):
    x = []
    y = []

    with open(filename,'r') as fp:
        plots = csv.reader(fp, delimiter=delimiter)
        # guess the labels
        if read_labels_from_first_line:
            labels = next(plots, None)
            if labels is not None:
                if xlabel is None:
                    xlabel = labels[xindex]
                if ylabel is None:
                    ylabel = labels[yindex]
        # read the data
        for row in plots:
            x.append(xtype(row[xindex]))
            y.append(ytype(row[yindex]))

    plt.plot(x, y)
    if title is not None:
        plt.title(title)
    if xlabel is not None:
        plt.xlabel(xlabel)
    if ylabel is not None:
        plt.ylabel(ylabel)
    #plt.legend()
    plt.show()
```

```python
plot_csv('buffered-report.csv', title="tokio::io::copy")
```

```python
plot_csv('basic-report.csv', title="basic")
```

```python
plot_csv('buffered-report.csv', title="tokio::io::copy", yindex=2)
```
